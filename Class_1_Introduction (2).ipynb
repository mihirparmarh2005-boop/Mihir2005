# Edit and Command Mode

```python
# - Every cell has edit and command mode.
# - We can write in a cell only when it is edit mode.
# - To convert a cell from command mode to edit mode, press enter or just click on the cell.

# - In command mode we can not write in the cell.
# - In command mode we can select multiple cells and perform some operation on all of them altogether.
# - In command mode we apply shortcut keys.
# - To convert a cell from edit mode to command mode, press esc key.
```

# Code and Markdown Editor

```python
# - A cell can act as a code editor or the text editor.
# - When in code mode then only Python code is expected.
# - If we write anything else then it throws errors.

# - When in markdown mode then a cell acts as a text editor.
```

# Shortcut Keys

```python
# - To exceute a cell -> Press shift + enter
# - To add a cell above a particular cell -> press a in command mode
# - To add a cell below a particular cell -> press b in command mode
# - To delete a cell -> Press x in command mode.
# - To convert a cell from code to markdown -> Press m in command mode.
# - To convert a cell from markdown to code -> Press y in command mode.
# - To undo -> Press z in command mode.
```
