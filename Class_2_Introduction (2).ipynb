```python
# - It is a small software that converts the HLL code into LLL code and vice-versa.
```

# Types of Translators

```python
# - 1- Compiler
# - 2- Interpreter
```

### 1- Compiler

```python
# - It reads the whole source code (HLL code) first then it converts the whole source code to machine code at one go.
```

### 2- Interpreter

```python
# - It reads the source code line by line and converts into machine code.
# - It reads and converts one line at a time.
```

### Compiled and Interpreted Languages

```python
# - Any language that uses compiler is a compiled language.
# Example - Java, C, C++ etc.

# - Any language that uses interpreter is an interpreted language.
# Example - Python, R, PHP, Ruby etc.
```

# Python is compiled or Interpreted?

### How a Python progra gets executed?

# Source Code

```python
# - The code that we write as a programmer is known as source code.
# - The file extension of the Python source code is .py.
```

### Intermediate Code

```python
# - When we run a Python program.
# - First compiler comes into  picture.
# - The compiler reads the whole source code first and if there is no error then it converts the source code into intermediate code.
# - If compiler catches any error in the source code then it does not convert it to intermediate code.
# - In this case we don't get any output.
# - The Python intermediate code is also known as bytecode.
# - The file extension of Python bytecode is .pyc.
```

### Machine Code

```python
# - As soon as Bytecode gets generated, interpreter comes into picture.
# - It reads the bytecode line by line and converts it into machine code.
# - If interpreter catches any error it stops the execution right away.
# - Until interpreter catches any error, we get the output.
```
